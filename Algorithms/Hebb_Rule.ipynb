```python
import numpy as np
import math

x1 = [1, -2, 1.5, 0]
x2 = [1, -0.5, -2, -1.5]
x3 = [0, 1, -1, 1.5]

w1 = [1, -1, 0, 0.5]
w1t = np.array(w1).T
net1 = np.sum(w1t*x1)
print('net1 = ',w1,'*',x1,' = ',net1)

w2 =[0,0,0,0]
for i in range(4):
    w2[i] = w1[i] + np.sign(net1)*x1[i]
print('w2 = ',w1,'+',x1,' = ',w2)

w2t = np.array(w2).T
net2 = np.sum(w2t*x2)
print('net2 = ',w2,'*',x2,' = ',net2)

w3 =[0,0,0,0]
for i in range(4):
    w3[i] = w2[i] + np.sign(net2)*x2[i]
print('w3 = ',w2,'-',x2,' = ',w3)

w3t = np.array(w3).T
net3 = np.sum(w3t*x3)
print('net3 = ',w3,'*',x3,' = ',net3)

w4 =[0,0,0,0]
for i in range(4):
    w4[i] = w3[i]  + np.sign(net3)*x3[i]
print('w4 = ',w3,'-',x3,' = ',w4)

def f(net):
    return round(2/(1+math.exp(-net))-1,3)

print('f(net1) = ',f(net1))
for i in range(4):
    w1[i] = w1[i] + f(net1)*x1[i]
print('w1 = ',w1) 

print('f(net2) = ',f(net2))
for i in range(4):
    w2[i] = w1[i] + f(net1)*x1[i]
print('w2 = ',w2)  

print('f(net3) = ',f(net3))
for i in range(4):
    w3[i] = w2[i] + f(net2)*x2[i]
print('w3 = ',w3)  
```
