```python
import numpy as np

#p=3
w1 = np.array([[65,70]])
w2 = np.array([[75,50]])
w3 = np.array([[110,55]])

x = np.array([[45,85],
            [50,43],
            [40,80],
            [55,42],
            [200,43],
            [48,40],
            [195,41],
            [43,87],
            [190,40]])


for i in range(9):
    norm1 = np.linalg.norm(x[i]-w1[i])
    print("||x-w1|| = ",norm1)
    norm2 = np.linalg.norm(x[i]-w2[i])
    print("||x-w2|| = ",norm2)
    norm3 = np.linalg.norm(x[i]-w3[i])
    print("||x-w3|| = ",norm3)
    m = min(norm1,norm2,norm3)
    
    if(m == norm1):
        print("w1",w1[i],"=> castigator! => Scaun")
        c = int(input("c="))
        a = [0,0]
        a = w1[i]+c*(x[i]-w1[i])
        w1 = np.append(w1,[a],axis = 0)
        w2 = np.append(w2,[w2[i]],axis = 0)
        w3 = np.append(w3,[w3[i]],axis = 0)
    elif(m == norm2):
        print("w2",w2[i],"=> castigator! => Masa")
        c = int(input("c="))
        a = [0,0]
        a = w2[i]+c*(x[i]-w2[i])
        w2 = np.append(w2,[a],axis = 0)
        w1 = np.append(w1,[w1[i]],axis = 0)
        w3 = np.append(w3,[w3[i]],axis = 0)
    else:
        print("w3",w3[i],"=> castigator! => Pat")
        c = int(input("c="))
        a = [0,0]
        a = w3[i]+c*(x[i]-w3[i])
        w3 = np.append(w3,[a],axis = 0)
        w1 = np.append(w1,[w1[i]],axis = 0)
        w2 = np.append(w2,[w2[i]],axis = 0)
    
```
